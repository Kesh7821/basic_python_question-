```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.
```

```python
# mutable
def modify_list(items):
    items[0] = "modified"
    print("Inside function - list:", items)
# immutable
def modify_number(num):
    num = num + 10
    print("Inside function - number:", num)
list = ["original", "value"]
print("Before function - list:",list)
modify_list(list)
print("After function - list:", list)
my_number = 5
print("Before function - number:", my_number)
modify_number(my_number)
print("After function - number:", my_number)
```

```python
#14. Write a program to reverse a string using a while loop
```

```python
rev_string=""
string=input("enter a string :")
index_num=len(string)-1
while index_num>=0:
    rev_string=rev_string+string[index_num]
    index_num=index_num-1

print("reversed string : ",rev_string)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop
```

```python
num=int(input("enter a number :"))
fact=1
while num>0:
    fact=fact*num
    num=num-1

print("factorial:",fact)
```
